For one of my models, I chose to use and tune a LightGBM model because of its efficiency with large datasets and how it can handle categorical features well. I broke up the modeling is 4 tuning stages to see how different hyperparameter tuning strategies compared. I first used a base model, and then applied GridSearchCV, RandomizedSearchCV, and Optuna. 

When evaluating the model performance on different methods of tuning, the base model was not the best performing, but after hyperparameter tuning, model performance increased. The base model still achieved an accuracy of 98.48% and a weighted F1-score of 0.98, however the tuned models specifically RandomizedSearchCV and Optuna pushed the model's performance even higher. The tuning process worked well at refining the model's predictions across the different classes. I used parameters like learning_rate, max_depth, num_leaves, and other parameters listed in other people's Kaggle notebooks, subsample and colsample_bytree. RandomizedSearchCV and Optuna models increased the overall weighted F1-score to 0.99. Tuning also improved the model's recall for the Medium class from 0.97 to 0.98 showing that the tuned models become better at identifying instances of medium irrigation need.

Although tuning worked well for majority of the dataset, it did not work as well for the High irrigation need class. Across all three tuning methods, the High class was stuck at 0.92 which means the model still misses 8% of the cases where high irrigation is required. Tuning alone was not enough to fix it as this is the most critical class to get right. This can mean the model struggled to recognize this pattern.

Another model I decided to use and tune was a CatBoost model as it worked to handle categorical features using ordered boosting. I also broke up the modeling in 4 stages with a base model, tuning with GridSearchCV, RandomizedSearchCV, and lastly Optuna. 

After evaluating the model results, the base model surprisingly performed the best. The model had an accuracy of 99% and a a weighted F1-Score of 0.98. It also had a Recall of 0.92 to predict the "High" irrigation need class. Through multiple steps of tuning the model, the performance would sometimes drop. I used parameters like tree depth, iterations, and learning_rate through GridSearchCV which made the model accuracy drop to 98% and the Recall for the High class to drop from 0.92 to 0.90. After tuning with RandomizedSearchCV and Optuna, the Recall did go back up to 0.92 like the base model, but neither was able to improve the accuracy any further.

For CatBoost, tuning the base model almost made small to no improvement to the metrics. Because of the lack of improvement, this can mean CatBoost's default hyperparameters can perform well on its own for datasets with a large amount of categorical features.

Like other tree-based models, CatBoost could not pass 0.92 Recall for the "High" irrigation need class. This class is critical to predict correctly and since tuning methods couldn't improve this metric, it shows that the model was not able to improve any further. To improve future modeling I would adjust the class weights, for example, penalize the model using Lasso or Ridge regression more for missing "High" predictions, or use feature engineering to create new interaction features, rather than only relying on hyperparameter tuning.

## LightGBM vs. CatBoost

How did the different models compare for the work?

LightGBM and CatBoost are both gradient boosting algorithms that build ensembles of decision trees, but they are different in how they handle data. LightGBM is ideal for speed and efficiency on large datasets, while CatBoost is actually designed to handle categorical features with very preprocessing and tuning, as we were able to see with the model results. Both models are important because they are meant to take structured, tabular data. 

Both models performed well overall, each had a high accuracy and weighted F1-Scores. LightGBM had small but some improvements through hyperparameter tuning with its F1-Score increasing from 0.98 to 0.99, which did improve the overall accuracy. CatBoost did perform the best even with just its default settings, performing well without the methods of hyperparameter tunings. After tuning three different ways (GridSearchCV, RandomizedSearchCV, and Optuna), they weren't any noticeable improvements that stood out apart from how the baseline model performed, and in some cases after tuning, it slightly reduced performance. Both models did struggle with the most critical classification of "high" irrigation need, where recall could not pass 0.92 across 4 stages.

For my leaderboard performance on Kaggle, I am currently in 1893th place. LightGBM outperformed CatBoost (0.95928 to 0.95809). Overall, LightGBM benefitted more from tuning and gave a better performance, while CatBoost was more stable and performed well without major hyperparameter tunings.
