3. 
My approach started with building a simple baseline model and moving towards more complex ensemble models. First, I made a baseline using Logistic Regression to get an initial sense of the dataset. Then, I tested an ensemble Bagging approach using a Random Forest Classifier, and then used a Boosting approach using XGBoost. 

What didn't work well: I ran into a major issue where my Random Forest model was taking a long time to run and overfitting. After debugging, I realized my model was accidently trying to predict the id column, and once I dropped the column from the features and only predicted on Irrigation_Need, the training became much faster and more accurate. I also had to reduce the size of the GridSearchCV parameter grid and lower the cross-validation folds because searching across too many deep decision tres on a 600,000+ row dataset was too heavy. 

What worked well: Reducing the size of the GridSearchCV and lowering cross-validation folds helped improve model performance and ensuring that it ran properly. I had subsequent improvements from hyperparameter tuning adjusting for n_estimators, max_depth, and min_samples_split using GridSearchCV. They were relatively small improvements but small adjustments gave a reasonably well enough score.

4. Bagging vs. Boosting

Bagging: In the Bagging model, it built decision trees to reduce the variance of the model and prevent overfitting. It was much easier to set up in code because it naturally accepted my dummy-encoded categorical variables and directly outputted my string lables of Low, Medium, High. 

Boosting: In the Boosting model, it built trees sequentially where each new tree tries to correct the errors (reduce bias) made by the previous trees. XGBoost was much stricter with the data preprocessing because troubleshooting required me to manually map my target categories to integers (0, 1, 2) and align the test set feature datatypes to match the training set features to avoid crashing the model. While XGBoost required more careful data formatting and troubleshooting, it got a higher final score on the Kaggle Leaderboard compared to the Random Forest model.
