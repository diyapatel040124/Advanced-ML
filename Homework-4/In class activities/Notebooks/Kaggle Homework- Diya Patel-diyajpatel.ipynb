## 1. Ensemble Evaluation and Performance
### Evaluate whether the ensemble improves performance compared to individual models and support your evaluation with specific results. 
The ensemble method improved overall predictive performance compared to the individual models slightly. The baseline CatBoost model achieved an accuracy of .9850 on the validation set and LightGBM achieved .9848. When using equal-weights probability averaging 50/50, the accuracy increased slightly to .9851, which means that stacking the two models helped to take out small errors. When using Optuna to find th best custom weights for probability averaging, the model performance was .9855. The Optuna tuning method assigned a higher weight to CatBoost .529 compared to .471. This makes sense because the CatBoost model was much stronger for its baseline model and had good probabilities which gave the best overall generalization. 

## 2. Feature Engineering Reflection 
### Implemented a mix of approaches to represent the data differently rahter than relying on the raw features.
Interactions: I created a Temp_Humidity_Index to create the relationship between temperature and humidity on evaporation and Rain_Temp_Ratio to represent rainfall relative to the temperature. I chose to interact humidity and rain with temperature because interacting weather conditions has likelihood of increasing model performance and can show the overall feature importance after interaction. 

Transformations: I used a Box-Cox transformation with Rainfall_BoxCox to handle the variance and right-skew of the rainfall data after taking a quick look at the EDA charts.

Clustering/Grouping: I also used K-Means clustering to create a Soil_Cluster feature based on soil metrics; like Soil_pH, Soil_Moisture, Organic_Carbon, Electrical_Conductivity

Encoding: Lastly, I applied count/frequency encoding for Crop_Type_Count and Region_Count to get a count for each of the variables.

## 3. What worked well and what didn't work well?
The interaction term of Temp_Humidity_Index and clustering of Soil_Cluster worked well because they made it easier to separate the Irrigation_Need classes with fewer tree splits. Count Encoding didn't get the best model improvements mostly because because by definition, boosting models like LightGBM and CatBoost already handle native categorical variables fairly well. 

## 4. Were model improvements meaningful or small?
Most of the model improvements were small. The base models were already predicting at a high accuracy of .98 and after ensembling pushed performance from .985 to .9855. Although this was an improvement to my last score in the Kaggle Competition leaderboard, the baseline models were already accurate enough. 

## 5. What will you use moving forward?
Moving forward I want to continue to use probability-averaged ensembles tuned with Optuna. It maximizes the strengths of LightGBM's histogram-based splitting and CatBoost's oblivious trees.
